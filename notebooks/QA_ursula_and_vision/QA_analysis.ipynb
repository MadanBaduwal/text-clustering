 # Preparing data to evaluate models

```python
import glob

files = glob.glob('*.csv')  
```

```python
import pandas as pd

frames = []
for file in files:
    df = pd.read_csv(file)
    frames.append(df)
```

```python
data = pd.concat(frames) # stack dataframe
data = data.reset_index(drop=True) # reset index dataframe 
```

```python
import numpy as np

data['manual'].replace(' ', np.nan, inplace=True)

data= data.dropna(subset=['manual'])
data = data.reset_index(drop=True) # reset_index after drop column
```

```python
data.tail()
```

```python
data["manual"].unique()
```

```python
data["manual"].replace({"Np": "NP", "p": "P"}, inplace=True)
```

```python
data['ursula_score'] = data.apply(lambda x: 'NP' if x['ursula_score'] <= 50 
               else ('P' if x['ursula_score'] >= 65 else 'C'), axis=1)
```

```python
data['vision_score'] = data.apply(lambda x: 'NP' if x['vision_score'] <= 0.50 
               else ('P' if x['vision_score'] >= 0.65 else 'C'), axis=1)
```

```python
data.head(20)
```

```python
data.tail(20)
```

```python
data['vision_score'].unique()
```

```python
data = data[data['ursula_score'] != 'C']
data = data[data['vision_score'] != 'C' ]
data = data[data['manual'] != 'C' ]
data = data[data['manual'] != 'D' ]

data = data.reset_index(drop=True) # reset index dataframe 
```

# Deployed model accuracy (Ursula)

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns


cm = confusion_matrix(data['manual'],data['ursula_score'])
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);
```

```python
print(classification_report(data['manual'],data['ursula_score']))
```

# Training model accuracy(Wvision)

```python
cm = confusion_matrix(data['manual'],data['vision_score'])
ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);
```

```python
print(classification_report(data['manual'],data['vision_score']))
```
